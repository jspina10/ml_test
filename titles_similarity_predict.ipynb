```python
import pandas as pd
import nltk
from scipy.spatial import distance
from gensim.models import Word2Vec
```

```python
nltk.download('punkt')
```

## Ejecución
Ejecute esta celda para averiguar los titulos similares.

Debe ingresar un índice del dataframe, un número entero entre 0 y 9999.

```python
# Cargar el modelo Word2Vec previamente entrenado
model = Word2Vec.load("modelo.w2v")

# Cargar el dataset "item_titles_test.csv"
df = pd.read_csv("items_titles_test.csv")
# Tokenizar los títulos del dataset
df['ITE_ITEM_TITLE_tokenized'] = df['ITE_ITEM_TITLE'].apply(lambda x: nltk.word_tokenize(x.lower()))

# Definir la función para calcular la similitud entre dos títulos
def calculate_similarity(title1, title2):
    title1_embeddings = []
    for token in title1:
        if token in model.wv.vocab:
            title1_embeddings.append(model[token])
    title2_embeddings = []
    for token in title2:
        if token in model.wv.vocab:
            title2_embeddings.append(model[token])
    if len(title1_embeddings) == 0 or len(title2_embeddings) == 0:
        return 0.0
    
    title1_embeddings_mean = sum(title1_embeddings) / len(title1_embeddings)
    title2_embeddings_mean = sum(title2_embeddings) / len(title2_embeddings)

    cosine_distance = distance.cosine(title1_embeddings_mean, title2_embeddings_mean)
    similarity_score = 1 - cosine_distance

    return similarity_score

# Calcular la similitud para todos los títulos del dataset
index = input('ingrese el index del título en el df que quiere buscar:')
title1 = df['ITE_ITEM_TITLE_tokenized'][int(index)]
similarity_scores = []
row_scores = []
for title2 in df['ITE_ITEM_TITLE_tokenized']:
    score = calculate_similarity(title1, title2)
    row_scores.append(score)
    

results_df ={'ITE_ITEM_TITLE':[df['ITE_ITEM_TITLE'][int(index)] for i in range(len(df['ITE_ITEM_TITLE']))], 'ITE_ITEM_TITLE_2':list(df['ITE_ITEM_TITLE']), 'SCORE':row_scores}
result = pd.DataFrame(results_df).sort_values(by='SCORE', ascending=False)
```
